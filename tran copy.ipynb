```python
import pandas as pd

df=pd.read_csv('D:/cedt/TCAS67_maxmin - Sheet2.csv')
```

```python
df.head(10)
```

```python
# Extract major from หลักสูตร column
df['สาขา'] = df['หลักสูตร'].str.extract(r'สาขาวิชา(.*?)(?:\s|$)', expand=False)

# Clean up the extracted major
df['สาขา'] = df['สาขา'].str.strip()

# Fill NaN values with original หลักสูตร value where extraction failed
df.loc[df['สาขา'].isna(), 'สาขา'] = df.loc[df['สาขา'].isna(), 'หลักสูตร']

# Display the first few rows to verify
df[['หลักสูตร', 'สาขา']].head()
```
